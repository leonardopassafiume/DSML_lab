```python
import numpy as np
data = np.genfromtxt('mnist_test.csv', delimiter=',')
```

```python
y = data[:,0]
X = data[:,1:]
```

```python
image = X[0].reshape(28,28)
final_image = np.empty_like(image, dtype = str)
mask = image < 64
mask_2 = (image >= 64) & (image < 128)
mask_3 = (image >= 128) & (image < 192)
mask_4 = (image >= 192) & (image < 256)
final_image[mask] = ' '
final_image[mask_2] = '.'
final_image[mask_3] = '*'
final_image[mask_4] = '#'

np.set_printoptions(linewidth=200)
print(final_image)
```
